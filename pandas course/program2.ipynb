```python
import pandas as pd

salary = [25000, 32000, 28000, 40000, 50000]
emp = ['A', 'B', 'C', 'D', 'E']

# 1. Pehle series banayein
s = pd.Series(salary, index=emp)

# 2. Ise float mein badlein taaki 10% calculation sahi ho (V.Imp Step)
s = s.astype(float)

# 3. Increment 10% (< 30000)
s.loc[s < 30000] = s.loc[s < 30000] * 1.10

# 4. Decrease 5% (> 45000)
s.loc[s > 45000] = s.loc[s > 45000] * 0.95

print("Updated Salaries:")
print(s)

print("\nTotal Salary Expense:")
print(s.sum())
```
