```python
import pandas as pd
marks = [45, 67, 89, 34, 90, 55, 72]
subjects = ['Math', 'Physics', 'Chemistry', 'English', 'CS', 'Biology', 'History']
s = pd.Series(marks, index=subjects)

# 1. Marks > 70
print("Subjects > 70:\n", s[s > 70])

# 2. Above Average Count
avg = s.mean()
print("\nAbove Average Count:", (s > avg).sum())

# 3. Replace < 40 with Fail
s1 = s.copy().astype(object)
s1[s1 < 40] = "Fail"
print("\nAfter Replacing:\n", s1)
```
