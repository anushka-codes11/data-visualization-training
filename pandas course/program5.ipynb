```python
import pandas as pd

s1 = pd.Series([10, 20, 30], index=['A', 'B', 'C'])
s2 = pd.Series([5, 15, 25], index=['B', 'C', 'D'])

# 1. Normal addition (NaN dikhega)
res = s1 + s2
print("Addition with NaN:\n", res)

# 2. NaN ko 0 se replace karke add karna
res_fixed = s1.add(s2, fill_value=0)
print("\nAddition after replacing NaN with 0:\n", res_fixed)
```

```python
scores = [88, 92, 79, 93, 85]
players = ['P1','P2','P3','P4','P5']
s = pd.Series(scores, index=players)

# 1. Rank players (Descending order mein)
ranks = s.rank(ascending=False)
print("Ranks:\n", ranks)

# 2. Top 3 players
print("\nTop 3 Players:\n", s.sort_values(ascending=False).head(3))

# 3. Lowest score player
print("\nLowest Score Player:", s.idxmin(), "with score:", s.min())
```

```python
names = ['anuj', 'rahul', 'sneha', 'kiran', 'amit']
s = pd.Series(names)

# 1. Uppercase
print("Uppercase Names:\n", s.str.upper())

# 2. Names containing letter 'a'
print("\nNames with letter 'a':\n", s[s.str.contains('a')])

# 3. Replace 'anuj'
s_new = s.replace('anuj', 'Anuj Kumar')
print("\nAfter replacement:\n", s_new)
```

```python
import pandas as pd

data = [5, None, 15, 25, None, 35, 45]
s = pd.Series(data)

# Chained steps: Fill Median -> Filter > 20 -> Sort Descending -> Reset Index
result = (s.fillna(s.median())
           [lambda x: x > 20]
           .sort_values(ascending=False)
           .reset_index(drop=True))

print("Final Processed Series:\n", result)
```

```python
sales = [200, 450, 300, 150, 500, 700, 100]
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
s = pd.Series(sales, index=days)

# 1. Best and Worst
print(f"Best: {s.idxmax()} ({s.max()}), Worst: {s.idxmin()} ({s.min()})")

# 2. Total and Average
print(f"Total: {s.sum()}, Average: {s.mean()}")

# 3. 20% Bonus where sales > 400
s.loc[s > 400] = s[s > 400] * 1.20

# 4. Sort descending
print("\nFinal Sales (Descending):\n", s.sort_values(ascending=False))
```
